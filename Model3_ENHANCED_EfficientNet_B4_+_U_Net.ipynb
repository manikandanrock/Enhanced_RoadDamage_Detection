```python
# ==============================================================
# MODEL 3 ‚Äî ENHANCED EfficientNet-B4 + U-Net with ADVANCED Techniques
# ==============================================================

!pip install -q kaggle timm segmentation-models-pytorch albumentations==1.3.0 torch torchvision matplotlib tqdm
!pip install -q wandb tensorboardX

import os, json, shutil, glob, csv, random
from pathlib import Path
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.utils.tensorboard import SummaryWriter
import albumentations as A
from albumentations.pytorch import ToTensorV2
import segmentation_models_pytorch as smp

# --------------------------------------------------------------
# 1Ô∏è‚É£ Kaggle Auth & Dataset Download
# --------------------------------------------------------------
os.makedirs("/root/.kaggle", exist_ok=True)
if os.path.exists("/root/.kaggle/kaggle.json"):
    os.chmod("/root/.kaggle/kaggle.json", 0o600)
    print("‚úÖ Kaggle API configured (from uploaded kaggle.json)")
else:
    print("‚öôÔ∏è Manual Kaggle setup")
    KAGGLE_USERNAME = "manikandanvistas"
    KAGGLE_KEY = "7d5e147c041cfd67317ae1ec2a5e38fc"
    with open("/root/.kaggle/kaggle.json", "w") as f:
        json.dump({"username": KAGGLE_USERNAME, "key": KAGGLE_KEY}, f)
    os.chmod("/root/.kaggle/kaggle.json", 0o600)
    print("‚úÖ Kaggle API configured manually")

# --------------------------------------------------------------
# 2Ô∏è‚É£ Dataset Download & Structure
# --------------------------------------------------------------
DATA_ROOT = "/content/datasets"
os.makedirs(DATA_ROOT, exist_ok=True)

DATASET = "farzadnekouei/pothole-image-segmentation-dataset"
print(f"üì¶ Downloading dataset '{DATASET}'...")
!kaggle datasets download -d {DATASET} -p {DATA_ROOT} --unzip -q

src = f"{DATA_ROOT}/pothole-image-segmentation-dataset"
dst = f"{DATA_ROOT}/Pothole_Segmentation_YOLOv8"
if os.path.exists(src) and not os.path.exists(dst):
    shutil.move(src, dst)
print("‚úÖ Dataset ready at:", dst)

# --------------------------------------------------------------
# 3Ô∏è‚É£ PROPER Polygon ‚Üí Mask Conversion
# --------------------------------------------------------------
TRAIN_IMG = f"{dst}/train/images"; TRAIN_LBL = f"{dst}/train/labels"
VAL_IMG   = f"{dst}/valid/images"; VAL_LBL   = f"{dst}/valid/labels"

UNIFIED_SEG = f"{DATA_ROOT}/Unified_SegRoad"
for split in ["train", "valid"]:
    os.makedirs(f"{UNIFIED_SEG}/{split}/images", exist_ok=True)
    os.makedirs(f"{UNIFIED_SEG}/{split}/masks", exist_ok=True)

def yolo_polygon_to_mask(label_file, shape):
    """Convert YOLO polygon format to precise segmentation mask"""
    h, w = shape
    mask = np.zeros((h, w), np.uint8)
    if not os.path.exists(label_file):
        return mask

    with open(label_file, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) < 5:
                continue
            # Convert normalized polygon coordinates to pixel coordinates
            coords = np.array(list(map(float, parts[1:])), dtype=np.float32).reshape(-1, 2)
            coords[:, 0] *= w  # x coordinates
            coords[:, 1] *= h  # y coordinates
            pts = np.round(coords).astype(np.int32)
            if len(pts) >= 3:
                cv2.fillPoly(mask, [pts], 255)
    return mask

print("üîÅ Generating Unified Segmentation Dataset with POLYGON masks...")
for split, img_dir, lbl_dir in [("train", TRAIN_IMG, TRAIN_LBL), ("valid", VAL_IMG, VAL_LBL)]:
    for img_path in tqdm(sorted(glob.glob(f"{img_dir}/*.jpg")), desc=f"{split}"):
        img = cv2.imread(img_path)
        if img is None:
            continue
        h, w = img.shape[:2]
        label_path = os.path.join(lbl_dir, os.path.basename(img_path).replace(".jpg", ".txt"))
        mask = yolo_polygon_to_mask(label_path, (h, w))

        out_img  = f"{UNIFIED_SEG}/{split}/images/{os.path.basename(img_path)}"
        out_mask = f"{UNIFIED_SEG}/{split}/masks/{os.path.basename(img_path).replace('.jpg', '.png')}"
        shutil.copyfile(img_path, out_img)
        cv2.imwrite(out_mask, mask)

print("‚úÖ Unified dataset at:", UNIFIED_SEG)

# --------------------------------------------------------------
# 4Ô∏è‚É£ Data Quality Verification
# --------------------------------------------------------------
def verify_mask_conversion(dataset_path, num_samples=3):
    """Verify that polygon mask conversion is working properly"""
    img_dir = f"{dataset_path}/train/images"
    mask_dir = f"{dataset_path}/train/masks"

    img_files = sorted(glob.glob(f"{img_dir}/*.jpg"))[:num_samples]

    fig, axes = plt.subplots(num_samples, 2, figsize=(8, 3*num_samples))

    for i, img_path in enumerate(img_files):
        # Load image
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Load mask
        mask_path = img_path.replace('images', 'masks').replace('.jpg', '.png')
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

        axes[i, 0].imshow(img)
        axes[i, 0].set_title(f"Image: {os.path.basename(img_path)}")
        axes[i, 0].axis('off')

        axes[i, 1].imshow(mask, cmap='gray')
        axes[i, 1].set_title(f"Mask (sum: {mask.sum()})")
        axes[i, 1].axis('off')

    plt.tight_layout()
    plt.show()

print("üîç Verifying polygon mask conversion...")
verify_mask_conversion(UNIFIED_SEG)

# --------------------------------------------------------------
# 5Ô∏è‚É£ ENHANCED CONFIGURATION
# --------------------------------------------------------------
IMG_SIZE = 320  # Increased from 256 for more detail
BATCH_SIZE = 8  # Adjusted for memory constraints
EPOCHS = 50  # More epochs with better regularization
LR = 3e-4  # Slightly higher learning rate
WEIGHT_DECAY = 1e-5  # Lower weight decay
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
ENCODER = "timm-efficientnet-b4"  # More powerful encoder

# Advanced training parameters
ACCUMULATION_STEPS = 2  # Gradient accumulation
USE_AMP = True  # Mixed precision training
EARLY_STOPPING_PATIENCE = 15

DST_MODELS = "/content/models"
PRED_DIR = "/content/predictions"
CSV_LOG = "logs/model3_advanced_metrics.csv"

os.makedirs(DST_MODELS, exist_ok=True)
os.makedirs(PRED_DIR, exist_ok=True)
os.makedirs("logs", exist_ok=True)

# --------------------------------------------------------------
# 6Ô∏è‚É£ DATASET CLASS WITH ENHANCED AUGMENTATIONS
# --------------------------------------------------------------
class SegDataset(Dataset):
    def __init__(self, img_dir, mask_dir, augment=None):
        self.imgs = sorted(glob.glob(f"{img_dir}/*.jpg"))
        self.mask_dir = mask_dir
        self.augment = augment

    def __len__(self):
        return len(self.imgs)

    def __getitem__(self, idx):
        img_path = self.imgs[idx]
        mask_path = os.path.join(self.mask_dir, os.path.basename(img_path).replace('.jpg', '.png'))

        img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

        if mask is None:
            mask = np.zeros(img.shape[:2], np.uint8)

        if self.augment:
            aug = self.augment(image=img, mask=mask)
            img, mask = aug["image"], aug["mask"]

        # Convert to tensor and normalize
        img = img.float()
        mask = mask.float().unsqueeze(0) / 255.0
        return img, mask

def get_advanced_augmentations(img_size):
    train_tfms = A.Compose([
        A.Resize(img_size, img_size),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.2),
        A.RandomRotate90(p=0.3),
        A.ShiftScaleRotate(
            shift_limit=0.1,
            scale_limit=0.2,
            rotate_limit=25,  # Increased rotation
            border_mode=cv2.BORDER_CONSTANT,
            value=0,
            mask_value=0,
            p=0.8
        ),
        A.RandomBrightnessContrast(
            brightness_limit=0.3,
            contrast_limit=0.3,
            p=0.6
        ),
        A.HueSaturationValue(
            hue_shift_limit=10,
            sat_shift_limit=20,
            val_shift_limit=10,
            p=0.5
        ),
        A.GaussianBlur(blur_limit=5, p=0.3),
        A.MotionBlur(blur_limit=5, p=0.2),
        A.GaussNoise(var_limit=(10.0, 50.0), p=0.3),
        A.RandomGamma(gamma_limit=(80, 120), p=0.3),
        A.CoarseDropout(
            max_holes=8,
            max_height=img_size//8,
            max_width=img_size//8,
            min_holes=1,
            fill_value=0,
            p=0.3
        ),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])

    val_tfms = A.Compose([
        A.Resize(img_size, img_size),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])

    return train_tfms, val_tfms

train_tfms, val_tfms = get_advanced_augmentations(IMG_SIZE)

# Create datasets and dataloaders
train_ds = SegDataset(f"{UNIFIED_SEG}/train/images", f"{UNIFIED_SEG}/train/masks", train_tfms)
val_ds = SegDataset(f"{UNIFIED_SEG}/valid/images", f"{UNIFIED_SEG}/valid/masks", val_tfms)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print(f"üì∏ Train samples = {len(train_ds)} | Val samples = {len(val_ds)}")

# --------------------------------------------------------------
# 7Ô∏è‚É£ ENHANCED MODEL ARCHITECTURE
# --------------------------------------------------------------
class AdvancedUNet(nn.Module):
    def __init__(self, encoder_name="timm-efficientnet-b4", classes=1):
        super().__init__()
        self.unet = smp.Unet(
            encoder_name=encoder_name,
            encoder_weights="imagenet",
            in_channels=3,
            classes=classes,
            activation=None,
            decoder_attention_type="scse",  # Spatial and Channel SE attention
            decoder_use_batchnorm=True,
            decoder_channels=(256, 128, 64, 32, 16),  # Custom decoder channels
        )

    def forward(self, x):
        return self.unet(x)

model = AdvancedUNet(ENCODER).to(DEVICE)
print(f"‚úÖ Model created with {ENCODER} encoder")

# --------------------------------------------------------------
# 8Ô∏è‚É£ ADVANCED LOSS FUNCTIONS
# --------------------------------------------------------------
class FocalDiceLoss(nn.Module):
    def __init__(self, alpha=0.8, gamma=2.0, smooth=1e-6):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.smooth = smooth

    def focal_loss(self, pred_logits, target):
        bce = F.binary_cross_entropy_with_logits(pred_logits, target, reduction='none')
        probs = torch.sigmoid(pred_logits)
        p_t = probs * target + (1 - probs) * (1 - target)
        focal_weight = (1 - p_t) ** self.gamma
        return (self.alpha * focal_weight * bce).mean()

    def dice_loss(self, pred_logits, target):
        probs = torch.sigmoid(pred_logits)
        num = (probs * target).sum(dim=(1,2,3))
        den = probs.sum(dim=(1,2,3)) + target.sum(dim=(1,2,3))
        return (1 - (2*num + self.smooth)/(den + self.smooth)).mean()

    def tversky_loss(self, pred_logits, target, alpha=0.7, beta=0.3):
        probs = torch.sigmoid(pred_logits)
        true_pos = (probs * target).sum(dim=(1,2,3))
        false_neg = ((1 - probs) * target).sum(dim=(1,2,3))
        false_pos = (probs * (1 - target)).sum(dim=(1,2,3))
        return 1 - ((true_pos + self.smooth) /
                   (true_pos + alpha * false_neg + beta * false_pos + self.smooth)).mean()

    def forward(self, pred_logits, target):
        focal = self.focal_loss(pred_logits, target)
        dice = self.dice_loss(pred_logits, target)
        tversky = self.tversky_loss(pred_logits, target)
        return 0.4 * focal + 0.4 * dice + 0.2 * tversky

class MultiScaleLoss(nn.Module):
    def __init__(self, scales=[1.0, 0.5, 0.25]):
        super().__init__()
        self.scales = scales
        self.focal_dice = FocalDiceLoss()

    def forward(self, pred_logits, target):
        total_loss = 0.0
        for scale in self.scales:
            if scale != 1.0:
                # Downsample for multi-scale
                h, w = int(target.shape[2] * scale), int(target.shape[3] * scale)
                pred_resized = F.interpolate(pred_logits, size=(h, w), mode='bilinear', align_corners=False)
                target_resized = F.interpolate(target, size=(h, w), mode='nearest')
            else:
                pred_resized = pred_logits
                target_resized = target

            total_loss += self.focal_dice(pred_resized, target_resized)

        return total_loss / len(self.scales)

criterion = MultiScaleLoss()

# --------------------------------------------------------------
# 9Ô∏è‚É£ ENHANCED OPTIMIZER & SCHEDULER
# --------------------------------------------------------------
optimizer = optim.AdamW(
    model.parameters(),
    lr=LR,
    weight_decay=WEIGHT_DECAY,
    betas=(0.9, 0.999),
    eps=1e-8
)

# OneCycle learning rate policy
scheduler = optim.lr_scheduler.OneCycleLR(
    optimizer,
    max_lr=LR,
    epochs=EPOCHS,
    steps_per_epoch=len(train_loader) // ACCUMULATION_STEPS,
    pct_start=0.1,
    div_factor=10.0,
    final_div_factor=100.0
)

# --------------------------------------------------------------
# üîü ADVANCED METRICS
# --------------------------------------------------------------
def advanced_segmentation_metrics(pred_logits, target, threshold=0.5, eps=1e-6):
    probs = torch.sigmoid(pred_logits)
    preds = (probs > threshold).float()

    # Basic metrics
    intersection = (preds * target).sum(dim=(1,2,3))
    union = (preds + target - preds * target).sum(dim=(1,2,3))

    dice = (2 * intersection + eps) / (preds.sum(dim=(1,2,3)) + target.sum(dim=(1,2,3)) + eps)
    iou = (intersection + eps) / (union + eps)
    acc = (preds == target).float().mean(dim=(1,2,3))

    # Precision, Recall, F1
    true_pos = (preds * target).sum(dim=(1,2,3))
    false_pos = (preds * (1 - target)).sum(dim=(1,2,3))
    false_neg = ((1 - preds) * target).sum(dim=(1,2,3))

    precision = true_pos / (true_pos + false_pos + eps)
    recall = true_pos / (true_pos + false_neg + eps)
    f1 = 2 * (precision * recall) / (precision + recall + eps)

    # Specificity
    true_neg = ((1 - preds) * (1 - target)).sum(dim=(1,2,3))
    specificity = true_neg / (true_neg + false_pos + eps)

    return {
        'dice': dice.mean().item(),
        'iou': iou.mean().item(),
        'accuracy': acc.mean().item(),
        'precision': precision.mean().item(),
        'recall': recall.mean().item(),
        'f1': f1.mean().item(),
        'specificity': specificity.mean().item()
    }

# --------------------------------------------------------------
# 1Ô∏è‚É£1Ô∏è‚É£ ENHANCED TRAINING LOOP WITH MIXED PRECISION
# --------------------------------------------------------------
scaler = torch.cuda.amp.GradScaler() if USE_AMP else None

def train_epoch_advanced(model, loader, criterion, optimizer, device, accumulation_steps=1):
    model.train()
    running_loss = 0.0
    optimizer.zero_grad()

    for i, (imgs, masks) in enumerate(tqdm(loader, desc="Training")):
        imgs, masks = imgs.to(device, non_blocking=True), masks.to(device, non_blocking=True)

        # Mixed precision forward pass
        if USE_AMP:
            with torch.cuda.amp.autocast():
                preds = model(imgs)
                loss = criterion(preds, masks) / accumulation_steps

            scaler.scale(loss).backward()
        else:
            preds = model(imgs)
            loss = criterion(preds, masks) / accumulation_steps
            loss.backward()

        # Gradient accumulation
        if (i + 1) % accumulation_steps == 0:
            if USE_AMP:
                scaler.step(optimizer)
                scaler.update()
            else:
                optimizer.step()
            optimizer.zero_grad()

            # Update learning rate
            scheduler.step()

        running_loss += loss.item() * accumulation_steps

    return running_loss / len(loader)

def validate_epoch_advanced(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    metrics_accumulator = {
        'dice': 0.0, 'iou': 0.0, 'accuracy': 0.0,
        'precision': 0.0, 'recall': 0.0, 'f1': 0.0, 'specificity': 0.0
    }

    with torch.no_grad():
        for imgs, masks in tqdm(loader, desc="Validation"):
            imgs, masks = imgs.to(device, non_blocking=True), masks.to(device, non_blocking=True)

            if USE_AMP:
                with torch.cuda.amp.autocast():
                    preds = model(imgs)
                    loss = criterion(preds, masks)
            else:
                preds = model(imgs)
                loss = criterion(preds, masks)

            metrics = advanced_segmentation_metrics(preds, masks)

            running_loss += loss.item()
            for key in metrics_accumulator:
                metrics_accumulator[key] += metrics[key]

    # Average metrics
    avg_metrics = {key: value / len(loader) for key, value in metrics_accumulator.items()}
    return running_loss / len(loader), avg_metrics

# --------------------------------------------------------------
# 1Ô∏è‚É£2Ô∏è‚É£ TRAINING EXECUTION
# --------------------------------------------------------------
best_val_loss = float("inf")
best_metrics = {}
patience_counter = 0

# Initialize TensorBoard
writer = SummaryWriter('logs/tensorboard')

# Logging setup
with open(CSV_LOG, "w", newline="") as f:
    writer_csv = csv.writer(f)
    writer_csv.writerow(["epoch", "train_loss", "val_loss", "dice", "iou", "accuracy",
                        "precision", "recall", "f1", "specificity", "lr"])

print("üöÄ Starting Advanced Training...")
for epoch in range(1, EPOCHS + 1):
    # Train
    train_loss = train_epoch_advanced(
        model, train_loader, criterion, optimizer, DEVICE, ACCUMULATION_STEPS
    )

    # Validate
    val_loss, val_metrics = validate_epoch_advanced(model, val_loader, criterion, DEVICE)

    # Log to TensorBoard
    writer.add_scalar('Loss/Train', train_loss, epoch)
    writer.add_scalar('Loss/Validation', val_loss, epoch)
    for metric_name, metric_value in val_metrics.items():
        writer.add_scalar(f'Metrics/{metric_name}', metric_value, epoch)

    # Print comprehensive metrics
    lr_now = optimizer.param_groups[0]["lr"]
    print(f"\nEpoch {epoch:02d}/{EPOCHS}")
    print(f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | LR: {lr_now:.2e}")
    print("Validation Metrics:")
    for metric_name, metric_value in val_metrics.items():
        print(f"  {metric_name:12}: {metric_value:.4f}")

    # Log to CSV
    with open(CSV_LOG, "a", newline="") as f:
        writer_csv = csv.writer(f)
        writer_csv.writerow([epoch, train_loss, val_loss, val_metrics['dice'],
                           val_metrics['iou'], val_metrics['accuracy'], val_metrics['precision'],
                           val_metrics['recall'], val_metrics['f1'], val_metrics['specificity'], lr_now])

    # Save best model
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_metrics = val_metrics
        patience_counter = 0

        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'scheduler_state_dict': scheduler.state_dict(),
            'best_val_loss': best_val_loss,
            'best_metrics': best_metrics,
        }, f"{DST_MODELS}/model3_advanced_best.pth")

        print("üíæ Best model saved with metrics:")
        for k, v in best_metrics.items():
            print(f"    {k}: {v:.4f}")
    else:
        patience_counter += 1
        if patience_counter >= EARLY_STOPPING_PATIENCE:
            print(f"üõë Early stopping triggered after {epoch} epochs!")
            break

writer.close()

# Save final model
torch.save(model.state_dict(), f"{DST_MODELS}/model3_advanced_final.pth")
print("üíæ Final model saved!")

# --------------------------------------------------------------
# 1Ô∏è‚É£3Ô∏è‚É£ FINAL EVALUATION & VISUALIZATION
# --------------------------------------------------------------
print("\nüß™ Final Evaluation...")

# Load best model
checkpoint = torch.load(f"{DST_MODELS}/model3_advanced_best.pth")
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()

final_pred_dir = Path(PRED_DIR) / "final_advanced_results"
final_pred_dir.mkdir(parents=True, exist_ok=True)

# Final evaluation
total_metrics = {
    'dice': 0.0, 'iou': 0.0, 'accuracy': 0.0,
    'precision': 0.0, 'recall': 0.0, 'f1': 0.0, 'specificity': 0.0
}

with torch.no_grad():
    for imgs, masks in tqdm(val_loader, desc="Final Evaluation"):
        imgs, masks = imgs.to(DEVICE), masks.to(DEVICE)
        preds = model(imgs)
        metrics = advanced_segmentation_metrics(preds, masks)

        for key in total_metrics:
            total_metrics[key] += metrics[key]

# Calculate averages
final_metrics = {key: value / len(val_loader) for key, value in total_metrics.items()}

print("\nüìä COMPREHENSIVE FINAL RESULTS:")
print("=" * 50)
for metric_name, metric_value in final_metrics.items():
    print(f"  {metric_name:12}: {metric_value:.4f}")

print(f"\nüéØ Best achieved metrics during training:")
for metric_name, metric_value in best_metrics.items():
    print(f"  {metric_name:12}: {metric_value:.4f}")

# --------------------------------------------------------------
# 1Ô∏è‚É£4Ô∏è‚É£ VISUALIZATION
# --------------------------------------------------------------
print("\nüé® Generating Predictions Visualization...")
samples_to_show = 5
indices = random.sample(range(len(val_ds)), samples_to_show)

fig, axes = plt.subplots(samples_to_show, 4, figsize=(20, 5 * samples_to_show))

for idx_row, idx in enumerate(indices):
    # Get original image and mask
    original_img_path = val_ds.imgs[idx]
    original_mask_path = os.path.join(f"{UNIFIED_SEG}/valid/masks",
                                    os.path.basename(original_img_path).replace('.jpg', '.png'))

    # Load original image and mask
    original_img = cv2.cvtColor(cv2.imread(original_img_path), cv2.COLOR_BGR2RGB)
    original_gt_mask = cv2.imread(original_mask_path, cv2.IMREAD_GRAYSCALE)
    if original_gt_mask is None:
        original_gt_mask = np.zeros(original_img.shape[:2], np.uint8)

    # Prepare image for prediction
    img_for_pred = cv2.resize(original_img, (IMG_SIZE, IMG_SIZE))
    img_tensor = val_tfms(image=img_for_pred)['image'].unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        pred_logits = model(img_tensor)
        pred_prob = torch.sigmoid(pred_logits)[0, 0].cpu().numpy()
        pred_mask = (pred_prob > 0.5).astype(np.uint8) * 255

    # Resize prediction back to original size
    pred_mask_original_size = cv2.resize(pred_mask, (original_img.shape[1], original_img.shape[0]))
    pred_mask_original_size = (pred_mask_original_size > 127).astype(np.uint8) * 255

    # Create overlay
    overlay = original_img.copy()
    overlay[pred_mask_original_size == 255] = [255, 0, 0]  # Red for predictions

    # Add ground truth contours
    contours, _ = cv2.findContours(original_gt_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(overlay, contours, -1, (0, 255, 0), 3)  # Green for GT

    # Confidence map
    confidence_resized = cv2.resize(pred_prob, (original_img.shape[1], original_img.shape[0]))

    # Plot
    axes[idx_row, 0].imshow(original_img)
    axes[idx_row, 0].set_title("Original Image")
    axes[idx_row, 0].axis('off')

    axes[idx_row, 1].imshow(original_gt_mask, cmap='gray')
    axes[idx_row, 1].set_title("Ground Truth")
    axes[idx_row, 1].axis('off')

    axes[idx_row, 2].imshow(overlay)
    axes[idx_row, 2].set_title("Prediction (Red) + GT (Green)")
    axes[idx_row, 2].axis('off')

    conf_plot = axes[idx_row, 3].imshow(confidence_resized, cmap='hot', vmin=0, vmax=1)
    axes[idx_row, 3].set_title("Confidence Map")
    axes[idx_row, 3].axis('off')
    plt.colorbar(conf_plot, ax=axes[idx_row, 3], fraction=0.046)

    # Save individual result
    out_path = final_pred_dir / f"result_{idx:03d}.png"
    cv2.imwrite(str(out_path), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))

plt.tight_layout()
plt.show()

print(f"‚úÖ Saved prediction overlays to: {final_pred_dir}")
print("üéØ Advanced training and evaluation completed successfully!")
```

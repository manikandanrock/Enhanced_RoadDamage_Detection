```python
# ===============================================================
# AI-Powered Road Anomaly Segmentation (Res-UNet + DimAcc metric)
# ===============================================================

# ---------- STEP 1. INSTALL DEPENDENCIES ----------
!pip install -q kaggle timm albumentations torch torchvision tqdm matplotlib opencv-python

import os, json, glob, zipfile, shutil, random, cv2, numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
from PIL import Image
import albumentations as A
from albumentations.pytorch import ToTensorV2
import timm, matplotlib.pyplot as plt

# ---------- STEP 2. ENTER KAGGLE CREDENTIALS ----------
KAGGLE_USERNAME = "manikandanvistas"
KAGGLE_KEY      = "7d5e147c041cfd67317ae1ec2a5e38fc"

# ---------- STEP 3. AUTHENTICATE KAGGLE ----------
os.makedirs("/root/.kaggle", exist_ok=True)
with open("/root/.kaggle/kaggle.json","w") as f:
    json.dump({"username":KAGGLE_USERNAME,"key":KAGGLE_KEY},f)
os.chmod("/root/.kaggle/kaggle.json",600)
print("‚úÖ Kaggle API configured.")

# ---------- STEP 4. DOWNLOAD DATASET ----------
DATA_ROOT="/content/datasets"; os.makedirs(DATA_ROOT,exist_ok=True)
print("üì• Downloading Pothole Segmentation YOLOv8 dataset...")
!kaggle datasets download -d farzadnekouei/pothole-image-segmentation-dataset -p {DATA_ROOT} --unzip

matches=[p for p in glob.glob(f"{DATA_ROOT}/*") if "pothole" in p.lower()]
if not matches: raise FileNotFoundError("‚ùå Dataset folder not found.")
KAGGLE_DIR=matches[0]
print("‚úÖ Kaggle dataset located at:",KAGGLE_DIR)

# ---------- STEP 5. YOLO‚ÜíMASK CONVERSION ----------
def yolo_to_mask(img_path,label_path,mask_path):
    img=cv2.imread(img_path); h,w=img.shape[:2]
    mask=np.zeros((h,w),np.uint8)
    if not os.path.exists(label_path): return False
    for line in open(label_path):
        pts=np.array(line.strip().split()[1:],float).reshape(-1,2)
        pts[:,0]*=w; pts[:,1]*=h
        cv2.fillPoly(mask,[pts.astype(int)],255)
    cv2.imwrite(mask_path,mask); return True

MASK_OUT=os.path.join(KAGGLE_DIR,"masks_converted")
for split in ["train","valid"]:
    os.makedirs(f"{MASK_OUT}/{split}",exist_ok=True)
    print(f"üß© Converting {split} YOLO annotations ‚Üí binary masks...")
    for img in tqdm(glob.glob(f"{KAGGLE_DIR}/{split}/images/*")):
        b=os.path.splitext(os.path.basename(img))[0]
        yolo_to_mask(img,f"{KAGGLE_DIR}/{split}/labels/{b}.txt",
                     f"{MASK_OUT}/{split}/{b}.png")
print("‚úÖ Masks saved to:",MASK_OUT)

# ---------- STEP 6. MERGE DATASETS ----------
TARGET=f"{DATA_ROOT}/Unified_SegRoad"
for s in ["train","val","test"]:
    os.makedirs(f"{TARGET}/{s}/images",exist_ok=True)
    os.makedirs(f"{TARGET}/{s}/masks",exist_ok=True)

def add_dataset(img_dir,mask_dir):
    imgs=sorted(glob.glob(f"{img_dir}/*")); random.shuffle(imgs)
    n=len(imgs); n1,n2=int(.8*n),int(.9*n)
    splits={"train":imgs[:n1],"val":imgs[n1:n2],"test":imgs[n2:]}
    for sp in splits:
        for img in splits[sp]:
            b=os.path.splitext(os.path.basename(img))[0]
            m=None
            for e in [".png",".jpg",".jpeg"]:
                c=os.path.join(mask_dir,b+e)
                if os.path.exists(c): m=c; break
            if m:
                shutil.copy(img,f"{TARGET}/{sp}/images/{b}.jpg")
                shutil.copy(m,f"{TARGET}/{sp}/masks/{b}.png")

add_dataset(f"{KAGGLE_DIR}/train/images",f"{MASK_OUT}/train")
add_dataset(f"{KAGGLE_DIR}/valid/images",f"{MASK_OUT}/valid")
for fldr in ["SegRoad","SegDamage","SegRoadDamage"]:
    base=f"{DATA_ROOT}/{fldr}"
    if os.path.exists(base):
        imgs=glob.glob(f"{base}/**/images",recursive=True)
        msks=glob.glob(f"{base}/**/masks",recursive=True)
        if imgs and msks: add_dataset(imgs[0],msks[0])

for s in ["train","val","test"]:
    print(f"{s}: {len(os.listdir(f'{TARGET}/{s}/images'))} images")

# ---------- STEP 7. DATASET CLASS & AUGS ----------
class SegDataset(Dataset):
    def __init__(self,imgs,msks,tfm):
        self.imgs=sorted(glob.glob(f"{imgs}/*"))
        self.msks=sorted(glob.glob(f"{msks}/*"))
        self.tfm=tfm
    def __len__(self): return len(self.imgs)
    def __getitem__(self,i):
        im=np.array(Image.open(self.imgs[i]).convert("RGB"))
        mk=np.array(Image.open(self.msks[i]).convert("L")); mk=(mk>127).astype("float32")
        a=self.tfm(image=im,mask=mk); return a['image'],a['mask']

train_tf=A.Compose([A.Resize(256,256),A.HorizontalFlip(.5),
    A.RandomBrightnessContrast(.5),A.ShiftScaleRotate(.5),
    A.Normalize((.485,.456,.406),(.229,.224,.225)),ToTensorV2()])
val_tf=A.Compose([A.Resize(256,256),
    A.Normalize((.485,.456,.406),(.229,.224,.225)),ToTensorV2()])

train_ds=SegDataset(f"{TARGET}/train/images",f"{TARGET}/train/masks",train_tf)
val_ds=SegDataset(f"{TARGET}/val/images",f"{TARGET}/val/masks",val_tf)
train_dl=DataLoader(train_ds,batch_size=8,shuffle=True,num_workers=2)
val_dl=DataLoader(val_ds,batch_size=8,shuffle=False,num_workers=2)
print("üß© Train/Val:",len(train_ds),len(val_ds))

# ======================================================
# FINAL FIXED RESUNET (spatially aligned output)
# ======================================================
class ResUNet(nn.Module):
    def __init__(self, backbone='resnet34', pretrained=True):
        super().__init__()
        self.enc = timm.create_model(backbone, pretrained=pretrained, features_only=True)
        ch = self.enc.feature_info.channels()  # [64, 64, 128, 256, 512]

        # Decoder blocks (with proper concat handling)
        self.up1 = self._blk(ch[4] + ch[3], 256)
        self.up2 = self._blk(256 + ch[2], 128)
        self.up3 = self._blk(128 + ch[1], 64)
        self.up4 = self._blk(64 + ch[0], 32)

        self.out_conv = nn.Sequential(
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 1, kernel_size=1)
        )

    def _blk(self, in_c, out_c):
        return nn.Sequential(
            nn.Conv2d(in_c, out_c, 3, padding=1),
            nn.BatchNorm2d(out_c),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_c, out_c, 3, padding=1),
            nn.BatchNorm2d(out_c),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        feats = self.enc(x)
        x1, x2, x3, x4, x5 = feats  # [64,64,128,256,512]

        d4 = F.interpolate(x5, scale_factor=2, mode='bilinear', align_corners=False)
        d4 = self.up1(torch.cat([d4, x4], dim=1))

        d3 = F.interpolate(d4, scale_factor=2, mode='bilinear', align_corners=False)
        d3 = self.up2(torch.cat([d3, x3], dim=1))

        d2 = F.interpolate(d3, scale_factor=2, mode='bilinear', align_corners=False)
        d2 = self.up3(torch.cat([d2, x2], dim=1))

        d1 = F.interpolate(d2, scale_factor=2, mode='bilinear', align_corners=False)
        d1 = self.up4(torch.cat([d1, x1], dim=1))

        out = self.out_conv(d1)

        # ‚úÖ Ensure final output matches input spatial size exactly
        if out.shape[2:] != x.shape[2:]:
            out = F.interpolate(out, size=x.shape[2:], mode='bilinear', align_corners=False)

        return out


device="cuda" if torch.cuda.is_available() else "cpu"
model=ResUNet().to(device)

# ---------- STEP 9. LOSSES & METRICS ----------
def dice_loss(p,t,eps=1e-6):
    p=torch.sigmoid(p); inter=(p*t).sum((2,3)); union=p.sum((2,3))+t.sum((2,3))
    return 1-((2*inter+eps)/(union+eps)).mean()
def focal_loss(p,t,a=.8,g=2):
    p=torch.sigmoid(p); bce=F.binary_cross_entropy(p,t,reduction='mean'); pt=torch.exp(-bce)
    return a*(1-pt)**g*bce
def combined_loss(p,t): return dice_loss(p,t)+focal_loss(p,t)
def iou(p,t):
    p=(torch.sigmoid(p)>.5).float(); inter=(p*t).sum((1,2,3)); union=((p+t)>0).float().sum((1,2,3))
    return ((inter+1e-6)/(union+1e-6)).mean().item()
def dice_metric(p,t):
    p=(torch.sigmoid(p)>.5).float(); inter=(p*t).sum((1,2,3)); union=p.sum((1,2,3))+t.sum((1,2,3))
    return ((2*inter+1e-6)/(union+1e-6)).mean().item()
def dim_acc(p,t):
    p=(torch.sigmoid(p)>.5).float(); pa=p.sum((1,2,3)); ga=t.sum((1,2,3))
    r=torch.where(ga>0,pa/(ga+1e-6),torch.ones_like(ga)); return r.clamp(0,1).mean().item()

# ---------- STEP 10. TRAIN + LOG METRICS ----------
opt=torch.optim.SGD(model.parameters(),lr=.01,momentum=.9,weight_decay=1e-4)
sched=torch.optim.lr_scheduler.CosineAnnealingLR(opt,T_max=40)
best=0; hist={"IoU":[], "Dice":[], "DimAcc":[]}

for ep in range(1,41):
    model.train(); tl=0
    for img,msk in tqdm(train_dl,desc=f"Epoch {ep}/40"):
        img,msk=img.to(device),msk.to(device).unsqueeze(1)
        opt.zero_grad(); out=model(img); l=combined_loss(out,msk)
        l.backward(); opt.step(); tl+=l.item()*img.size(0)
    tl/=len(train_ds)
    model.eval(); vl=iou_s=dice_s=dim_s=0
    with torch.no_grad():
        for img,msk in val_dl:
            img,msk=img.to(device),msk.to(device).unsqueeze(1); out=model(img)
            vl+=combined_loss(out,msk).item()*img.size(0)
            iou_s+=iou(out,msk)*img.size(0); dice_s+=dice_metric(out,msk)*img.size(0); dim_s+=dim_acc(out,msk)*img.size(0)
    vl/=len(val_ds); iou_s/=len(val_ds); dice_s/=len(val_ds); dim_s/=len(val_ds); sched.step()
    hist["IoU"].append(iou_s); hist["Dice"].append(dice_s); hist["DimAcc"].append(dim_s)
    print(f"Epoch {ep:02d} | TrainLoss={tl:.4f} | ValLoss={vl:.4f} | IoU={iou_s:.4f} | Dice={dice_s:.4f} | DimAcc={dim_s:.4f}")
    if dim_s>best: best=dim_s; torch.save(model.state_dict(),"/content/best_resunet_dimacc.pth")

print("‚úÖ Training done | Best Dimensional Accuracy:",best)

# ---------- STEP 11. PLOT METRICS ----------
plt.figure(figsize=(10,4))
plt.plot(hist["IoU"],label="IoU"); plt.plot(hist["Dice"],label="Dice"); plt.plot(hist["DimAcc"],label="DimAcc")
plt.legend(); plt.title("Validation Metrics per Epoch"); plt.xlabel("Epoch"); plt.ylabel("Score"); plt.show()

# ---------- STEP 12. VISUALIZE PREDICTIONS ----------
model.eval(); samples=glob.glob(f"{TARGET}/val/images/*")[:6]
plt.figure(figsize=(12,8))
for i,pth in enumerate(samples):
    img=cv2.imread(pth)[...,::-1]; r=cv2.resize(img,(256,256))
    t=A.Normalize((.485,.456,.406),(.229,.224,.225))(image=r)
    ten=torch.tensor(t['image'].transpose(2,0,1)).unsqueeze(0).float().to(device)
    with torch.no_grad(): pr=torch.sigmoid(model(ten))[0,0].cpu().numpy()
    plt.subplot(2,3,i+1); plt.imshow(r); plt.imshow(pr>.5,alpha=.4,cmap='Reds'); plt.axis('off')
plt.suptitle("Res-UNet Predictions (Red = Damage Mask)",fontsize=15); plt.show()
```
